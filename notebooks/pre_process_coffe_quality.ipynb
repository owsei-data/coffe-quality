```python
import pandas as pd
data = pd.read_csv('coffee_eda.csv')
data.columns
```

```python
colunas_categoricas = [
    'Country of Origin',
    'Variety',
    'Processing Method'
]

#getdummies
data_final_modelo = pd.get_dummies(
    data,
    columns=colunas_categoricas,
    drop_first=True
)

data_final_modelo.info()
```

```python
data_modelo = data_final_modelo.drop(columns='Unnamed: 0')
data_modelo.to_csv('coffee_modelo.csv', index=False)
```
